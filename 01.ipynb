```python
y = df['RainTomorrow']
x = df.drop('RainTomorrow',axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
encoder=StandardScaler()
encoder.fit(x)
encoder=encoder.transform(x)
x_frame=pd.DataFrame(encoder,columns=x.columns)
```

```python
x_frame
```

### Split the data into train ,test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x_frame, y, random_state=124859 )
```

### Logistic RegressionCV

```python

def train_model_lg():
    from sklearn.linear_model import LogisticRegressionCV

    model = LogisticRegressionCV()
    model.fit(x_train, y_train)
    return model,"Lg"
```

### Navis Bayes

```python
def train_model_nb():
    from sklearn.naive_bayes import GaussianNB
    model=GaussianNB()
    model.fit(x_train,y_train)
    return model,"nb"
```

### Support Vector Machine

```python
def train_model_svm():
    from sklearn.svm import SVC
    model=SVC()
    model.fit(x_train,y_train)
    return model,"svm"
```

### KNneart Neghibour

```python
def train_model_KNN():
    from sklearn.neighbors import KNeighborsClassifier
    model=KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train,y_train)
    return model,"knn"
```

### Decission Tree

```python
def train_model_Dt():
    from sklearn.tree import DecisionTreeClassifier
    model=DecisionTreeClassifier(max_depth=3)
    model.fit(x_train,y_train)
    return model,"DT"
```

### Random Forest

```python
def train_model_randomforest():
    from sklearn.ensemble import RandomForestClassifier
    model=RandomForestClassifier(n_estimators=100)
    model.fit(x_train,y_train)
    return model,"random forest"
```

### Catboost

```python
def train_model_catboost():
    from catboost import CatBoostClassifier
    model=CatBoostClassifier()
    model.fit(x_train,y_train)
    return model,"catboost"
```

### AbaBOOST

```python
def train_model_adaboost():
    from sklearn.ensemble import AdaBoostClassifier
    model=AdaBoostClassifier()
    model.fit(x_train,y_train)
    return model,"adaboost"
    
```

### XGBoost

```python
def train_model_xgboost():
    from xgboost import XGBClassifier
    model=XGBClassifier()
    model.fit(x_train,y_train)
    return model,"XGBClassifier"
```

### Prepare the models list

```python
models=[train_model_lg(),train_model_nb(),train_model_svm(),train_model_KNN(),train_model_Dt(),train_model_randomforest(),train_model_catboost(),train_model_adaboost(), train_model_xgboost()]
```

### Evaluate the the model as per score

```python
def evaluate_model(model,model_name):
    from sklearn.metrics import  confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
    y_true=y_test
    y_predict=model.predict(x_test)
    a=accuracy_score(y_true,y_predict)*100
    p=precision_score(y_true,y_predict)*100
    r=recall_score(y_true,y_predict)*100
    f1=f1_score(y_true,y_predict)*100
    return model_name,a,p,r,f1
```

```python
rows=[]
for (model,model_name) in models:
    rows.append(evaluate_model(model,model_name))
```

```python
df2=pd.DataFrame(rows,columns=["Algorithm","accuracy_score","precision_score","recall_score","f1_score"])
df2
```
